```python
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

conn = sqlite3.connect("../db/baltasar.db")
def sql(query):
    #print(query)
    return pd.read_sql_query(query, conn)
```

```python
sql('''SELECT count(*) FROM Transfer''')
```

```python
sql('''SELECT 
    age, b.*
    FROM 
    EntityTaxonomy t
    INNER JOIN
    EntityMetadata b
    ON b.name = t.name
    WHERE age="week" and t.blocknumber = 7680001
    LIMIT 100''')
```

```python
dist = sql('''
SELECT 
    token, max(firstInTransfer) as lastCreated
FROM
    ETransfer
    INNER JOIN
    EntityMetadata
    ON name = ETransfer.`to`
GROUP BY token
''')
```

```python
dist['time'] = pd.to_datetime(dist["lastCreated"], unit='s')

sns.boxplot(dist['lastCreated'])
```

```python

dist.sort_values(by='time')
```

```python
sql('''
select 
    blocknumber / 100000 * 100000 as bl, count(*) as numTransfers
from 
ETransfer
GROUP BY bl
''')
```

```python
d = sql('''
SELECT
    blocknumber / 100000 * 100000 as periode, 
    count(distinct blocknumber) as numBlocks, 
    count(*) / 1.0 / count(distinct blocknumber) as transfers_per_block
FROM
    ETransfer
GROUP BY periode
''')
```

```python

sns.lineplot(data=d, x='periode', y='numBlocks')
plt.show()
sns.lineplot(data=d, x='periode', y='transfers_per_block')
```

```python
e = sql('''
SELECT
    blocknumber / 100000 * 100000 as periode, 
    count(distinct blocknumber) as numBlocks, 
    count(*) / 1.0 / count(distinct blocknumber) as transfers_per_block
FROM
    Transfer
WHERE not token = '0x'
GROUP BY periode
''')
d = sql('''
SELECT
    blocknumber / 100000 * 100000 as periode, 
    count(distinct blocknumber) as numBlocks, 
    count(*) / 1.0 / count(distinct blocknumber) as transfers_per_block
FROM
    ETransfer
WHERE not token = '0x'
GROUP BY periode
''')
```

```python

sns.lineplot(data=e, x='periode', y='numBlocks')
sns.lineplot(data=d, x='periode', y='numBlocks')
plt.show()
sns.lineplot(data=e, x='periode', y='transfers_per_block')
sns.lineplot(data=d, x='periode', y='transfers_per_block')
```
