A core problem for an exchange is to determine which user sent which transfer into the exchange. Two main approaches exist to solve this:
1. Use a smart contract and let users deposit using functions.
2. Generate a dedicated address for each event/user and let the user send money to this address, from where it is transferred to an address of the exchange.

Option two leaves a very distinctive trail. We can identify receiving addresses and asociated depositaddresses:
